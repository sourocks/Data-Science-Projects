Q.9 Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
intrn = requests.get('https://internshala.com/fresher-jobs')
intrn
```

```python
Soup = BeautifulSoup(intrn.content, 'lxml')
```

```python
jb = soup.find_all('div',class_='heading_4_5 profile')
intshp = soup.find_all('div',class_='internship_meta')
```

```python
job = []
company = []
CTC_apply = []
```

```python
for x in jb:
    for j in x.find_all('a'):
        job.append(j.text.strip().replace('\n',''))
for i in intshp:
    
    for c in i.find_all('a',class_='link_display_like_text'):
        company.append(c.text.strip().replace('\n',''))
    
    for t in i.find_all('div',class_='item_body'):
        CTC_apply.append(t.text.strip().replace('\n','').strip().replace('\n',''))
```

Separate CTC and Apply By Dates

```python
CTC = []
Apply = []
```

```python
for c in range (1,len(CTC_apply),3):
    CTC.append(CTC_apply[c])
    
for a in range(2,len(CTC_apply),3):
    Apply.append(CTC_apply[a])
```

```python
Apply
```

```python
Jobs = {'Title':job,'Company':company,'CTC(₹)':CTC, 'Apply By':Apply}
Jobs  = pd.DataFrame(Jobs)
Jobs
```

10. Write a python program to scrape house details from https://www.nobroker.in/ for any location. It should include house title, location, area, emi and price

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
hse = requests.get('https://www.nobroker.in/property/sale/mumbai/Marine%20Drive?searchParam=W3sibGF0IjoxOC45NDE0NTkyLCJsb24iOjcyLjgyMzYyOTUsInBsYWNlSWQiOiJDaElKSFZfLTNPSFI1enNSYWU3em11RUEzQzAiLCJwbGFjZU5hbWUiOiJNYXJpbmUgRHJpdmUifV0=&radius=2.0')
hse
```

```python
Soup = BeautifulSoup(hse.content)
```

```python
name = Soup.find_all('a',class_='nb__3CnI6')
loc = Soup.find_all('div',class_='nb__2CMjv')
Emi = Soup.find_all('div',class_='font-semi-bold heading-6')
ar = Soup.find_all('div',class_='nb__3oNyC')
pr = Soup.find_all('div',class_='font-semi-bold heading-6')
```

```python
Titles = []
Locations = []
EMI = []
Area = []
Price = []
```

```python
for i in name:
    for j in i.find_all('span'):
        Titles.append(j.text)
for l in loc:
    Locations.append(l.text)
    
for e in Emi:
    EMI.append(e.text)
for a in ar:
    Area.append(a.text)
for p in pr:
    for q in p.find_all('span'):
        Price.append(q.text)
```

Extract Emi and Price data

```python
emi = []
for i in range(1,len(EMI),3):
    emi.append(EMI[i])
```

```python
price = []
for p in range(1,len(Price),2):
    price.append(Price[p])
```

```python
House = {'Title':Titles,'Location':Locations,'Area':Area, 'EMI estimated':emi,'Price':price[:10]}

df = pd.DataFrame.from_dict(House, orient='index')

df
```
